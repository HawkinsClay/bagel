```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

fig, axes = plt.subplots( nrows=1, ncols=1, figsize=(40, 4) )
mylist = [ plt.imshow( mpimg.imread( 'slices/out_%s.png' % str(i+1) ) ) for i in range(140) ] 
df = pd.read_csv( 'sb_data.csv' )
df.index = df.sample
df = df.drop('sample', axis=1) 
#df = df.dropna()
df = df.T
mask = df.isnull()
sns.heatmap( df, square=True, mask=mask, xticklabels=mylist )
fig.show()
```

```python
from weblogolib import * 
for i in range(1, 455):
    with open( 'msa_aligned_to_bagel.fasta' ) as fin, open( 'slices/out_%d.png' % i , 'wb' ) as fn:
        seqs = read_seq_data(fin) 
        data = LogoData.from_seqs(seqs)
        options = LogoOptions( logo_margin=0.1, text_font="Helvetica", resolution=300, logo_start=i, logo_end=i, show_yaxis=False, show_xaxis=False, show_ends=False, show_fineprint=False, show_errorbars=False )
        format = LogoFormat(data, options)
        eps = png_print_formatter( data, format )
        fn.write( eps )
```

```python
import matplotlib.image as mpimg 
n = 140
fi, ax = plt.subplots(nrows=1, ncols = n, figsize=(n, 1) )
mylist = [ mpimg.imread( 'slices/out_%s.png' % str(i+1) ) for i in range(n) ] 
for i, j in enumerate( mylist ):
    plt.setp( ax[i].get_xticklabels(), visible=False )
    plt.setp( ax[i].get_yticklabels(), visible=False )
    ax[i].imshow(j)
    
fi.savefig( 'fi.png' )
```
