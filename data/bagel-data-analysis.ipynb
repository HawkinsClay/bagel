## Processing the raw data 

To begin, I wrote a script, `process.py`, to process the raw data. The script splits up the data by sample and attempts three fits on each sample.

$$y=mx+b \qquad kobs=\frac{kcat*S}{km+S} \qquad kobs=\frac{kcat*S}{km+S+ki}$$. 

`process.py` constructs a new data table, stored in `raw-out.csv`, keyed by sample name, and diagnostic plots for each sample.

```python
ro = pd.read_csv('out/raw-out.csv')
```

Let's add some useful metadata

```python
# metadata 
ro['seq'] = ro['sample'].str.replace(r'[a-z]', '')
ro['newaa'] = ro['sample'].str[-1]
ro['oldaa'] = ro['sample'].str[0]
```

Processing with `process.py` reveals that we have 121 unique sample names in our data set, with an average yield of 0.71 ± 0.45 mg/mL and a maximum recorded yield of 2 mg/mL. Of course, a few of these are wild type.

## Lots and lots of errors

As you can see from the summary table, the average error for kcat is actually larger than the magnitude of the value. Clearly, we have come cleaning up to do. 

## Cleanup 

First things first, let's filter out all the wild types and average them.

```python
import re
wts = ro[(ro['sample'].str.contains('wt'))] 
wts
```

```python
mean_wts = wts.mean()
mean_wts.set_value('sample', 'wt')
mean_wts
```

So now that we have the mean values for the WT, let's remove all the individual WTs from the data set and add back in our average value.

```python
ro = ro[( ro['sample'].str.match('wt') == False )]
ro.loc[107] = mean_wts
ro.tail()
len(ro)
```

This leaves us with 109 mutants tested. We're already getting close to "under 100".

## Linear fits

Probably a good place to start cleaning up is to determine what data fit best to straight lines. For this, let's say that anything with a linear r-square of greater than 0.95 is linear.

```python
linear = ro[(ro['R'] > 0.95)]
ro = ro[(ro['R'] < 0.95)]
len(linear)
```

```python
linear['eff'] = linear['slope']
linear['err3'] = linear['std_err']
linear['kcat'] = linear['km'] = linear['err1'] = linear['err2'] = 'NaN'
```

Right now, we're left with 80 mutants that will provide kcats and KMs. The ones that fail our error cutoff are

```python
ro = ro[(ro['err1']/ro['kcat']<0.25)|(ro['err2']/ro['km']<0.25)]
```

```python
high_err
```

```python
len(ro)
```

And we're left with 77 mutants that will provide kcats and KMs, and 12 that will provide a kcat/km value, for a total of 89. Let's calculate the kcat/KM for these too

```python
ro['eff'] = ro['kcat']/ro['km']
ro['err3'] = ro['kcat']/ro['km'] * ( (ro['err1']/ro['kcat'])**2 + (ro['err2']/ro['km'])**2 )**0.5

# logs and normalized logs
ro['log kcat'] = log(ro['kcat']+1)
ro['log km'] = log(ro['km']+1)
ro['normalized kcat'] = ro['log kcat'] / max(ro['log kcat'])
ro['normalized km'] = ro['log km'] / max(ro['log km'])
```
