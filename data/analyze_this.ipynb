```python
import pandas
from scipy.optimize import curve_fit
from scipy.stats import linregress
from numpy import linspace
import matplotlib.pyplot as plt
 
def mm(S, kcat, km): return kcat*S/(km+S)
def si(S, kcat, km, ki): return kcat*S/(km+S*(1+S/ki))
def f(m, x, b): return m*x+b
 
def fit(data):
  # linear params
  slope, intercept, r_value, p_value, std_err = linregress(data['s'], data['kobs'])
  
  # michaelis menten params
  p0 = ( data['kobs'].max(), data['s'].mean() )
  (kcat, km), cov = curve_fit(mm, data['s'], data['kobs'], p0=p0) 
  err1, err2 = [ abs(cov[i][i])**0.5 for i in range(2) ]
  
  # substrate inhibition params
  try:
    p0 = ( data['kobs'].max(), data['s'].mean(), data['s'].mean() )
    (si_kcat, si_km, ki), si_cov = curve_fit(si, data['s'], data['kobs'], p0=p0)
    err4, err5, err6 = [ abs(si_cov[i][i])**0.5 for i in range(3) ]
  except:
    si_kcat, si_km, ki, err4, err5, err6 = [ False ] * 6
 
  # splines for the various fits
  #mm_x = lm_x = si_x = linspace(0,max(data['s']))
  #mm_y = [mm(xx,kcat,km) for xx in mm_x]
  #si_y = [si(xx,si_kcat,si_km,ki) for xx in si_x]
  #lm_y = [f(slope,xx,intercept) for xx in lm_x]
  
  result = { 
    'yield': data['yield'].min(), 'x': data['s'], 'y': data['kobs'],
    'kcat': kcat, 'err1': err1, 'km': km, 'err2': err2, 
    'slope': slope, 'std_err': std_err, 
    'si kcat': si_kcat, 'si_km': si_km, 'ki':  ki, 'err4':  err4, 'err5':  err5, 'err6':  err6,
    #'mm_x': mm_x, 'mm_y': mm_y, 'lm_x': lm_x, 'lm_y': lm_y, 'si_x': si_x, 'si_y': si_y,
  }
    
  return pandas.Series(result)
 
# io
plates = pandas.read_csv( 'raw_data_edited.csv' )
fits = plates.groupby( by='sample' ).apply( fit )
```

### Diagnostic plots for each mutant (find in supplemental_fig/plots/)

```python
for sample, fit in fits.iterrows():

  #print('Plotting sample %s' % sample)
 
  fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(14,4))
  fig.suptitle( '%s (yield %1.2f mg/mL)' % (sample, fit['yield']), fontsize=18)
  fig.subplots_adjust(top=0.66) # make room for labels
  #fig.tight_layout()
 
# splines for the various fits
  mm_x = lm_x = si_x = linspace(0,max(fit['x']))
  
  mm_y = [mm(xx,fit.kcat,fit.km) for xx in mm_x]
  si_y = [si(xx,fit.kcat,fit.km,fit.ki) for xx in si_x]
  lm_y = [f(fit.slope,xx,0) for xx in lm_x]    
    
  axes[0].scatter(fit.x, fit.y)
  axes[0].plot(lm_x, lm_y)
  axes[0].set_title( 'Linear fit\nslope = %.0d ± %.0f' % ( fit.slope, fit.std_err) )
  axes[0].set_ylabel( 'Rate observed (1/min)' )
 
  axes[1].scatter(fit['x'], fit['y'])
  axes[1].plot(mm_x, mm_y)
  axes[1].set_title("Michaelis-Menten fit\nkcat=%2.0f ± %.0f%%\nkm = %1.4f ± %0.2f%%" % (fit['kcat'], fit['err1']/fit['kcat']*100, fit['km'], fit['err2']/fit['km']*100))
  axes[1].set_xlabel('4-nitrophenyl-ß-D-glucoside (M)') 
  
  if 0 < fit['ki'] < 100:
    axes[2].scatter(fit['x'], fit['y'])
    axes[2].plot( si_x, si_y )
    axes[2].set_title("Michaelis-Menten with substrate inhibition\nki = %1.4f ± %2.0f%%" % (fit['ki'], fit['err6']/fit['ki']*100),)
  else:
    axes[2].set_title("No substrate inhibition fit") 
    
 
  fig.savefig('plots/%s.pdf' % sample)
  plt.close()
```
